## ***1.4a Data Cleaning***

```python
# we change column names as they havent been named in an appropriate manner; therefore we shall rename to get some uniformity
columns = ['Address', 'Cars', 'Bluecar_counter', 'Utilib_counter', 'Utilib_1.4_counter', 'Charge_slots', 'Charging_Status', 'City', 'Displayed_comment', 'ID', 'Kind', 'Geo_point', 'Postal_code', 'Public_name', 'Rental_status', 'Scheduled_at', 'Slots', 'Station_type', 'Status', 'Subscription_status', 'year', 'month', 'day', 'hour', 'minute']
autolib.columns = columns
autolib.columns
```

```python
# let us find any null values in the dataset.
autolib.isnull().sum()
```

```python
# from the cell above we note that 2 columns are almost empty and therefore they are irrelevant in our dataset and if not dropped will lead to major inconsistencies.
# dropping columns displayed comment & scheduled at
autolib.drop(['Displayed_comment','Scheduled_at'],axis=1, inplace =True)
```

```python
# view if there are any null entries in our data set
autolib.isnull().sum()
```

```python
# finding duplicate values
autolib.duplicated()
```

```python
# from the cell above we can note that no duplicate values exist in our datasetexist
```

```python
# from viewing the datasets we can make an observation;that the columns cars & bluecar counter are similar but before we decide to drop the irrelevant columnn (cars) we must first prove this
autolib['Cars'] == autolib['Bluecar_counter']
```

```python
# our theory was true, we now drop the column cars as it is irrelevant
autolib.drop('Cars', axis=1, inplace=True)
```

```python
# we now change the values of the clolumn day to represent actual days of the week in a non-numerical form
autolib['day'].replace({1: "Monday", 2: "Tuesday", 3: "Wednesday", 4: "Thursday", 5: "Friday", 6: "Sartuday", 7: "Sunday"}, inplace=True)
autolib
```

```python
# Question 1: Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.
autolib1 = autolib[(autolib['City'] == 'Paris')]
autolib1
# we group the data: groupby hour then groupby bluecar counter
autolib1 = autolib1.groupby(['hour'])['Bluecar_counter'].max()
autolib1.sort_values(ascending = False)

# we can observe that the most popular hour and number of blue cars picked is the third hour; 7  bluecars
```

```python
# What is the most popular hour for returning bluecar cars?

# the question asks us to find out the most popular hour for returning all cars
# we will start with the bluecar
autolib1.loc[autolib1 == autolib1.min()]
autolib1.head(1)
```

```python
# most popular hour for returning utilib cars
autolib1 = autolib[(autolib['City'] == 'Paris')]
autolib1
# groupby hour then groupby utilib counter
autolib1 = autolib1.groupby(['hour'])['Utilib_counter'].max()
autolib1.sort_values(ascending = False)

# getting the highest hour
autolib1.loc[autolib1 == autolib1.min()]
autolib1.head(1)
```

```python
# most popular hour for returning utilib 1.4 cars

# most popular hour for returning utilib 1.4 cars
autolib1 = autolib[(autolib['City'] == 'Paris')]
autolib1
# groupby hour then groupby utilib 1.4 counter
autolib1 = autolib1.groupby(['hour'])['Utilib_1.4_counter'].max()
autolib1.sort_values(ascending = False)

# getting the highest hour
autolib1.loc[autolib1 == autolib1.min()]
autolib1.head(1)
```
